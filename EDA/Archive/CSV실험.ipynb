```python
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab

Xs = [i for i in range(2000, 2022)]
#ys = transposed_df_Cambodia[nonNullIndicators].T.loc["Population growth (annual %)"].tolist()
# NaN values are recorded as "..", so convert it to 0
ys = []
for i in range (0,9):
    ys.append(df_sample_1.loc[sea_countries[i]].tolist()[39:61])

for num in range (9):
    plt.title("sea_countries")
    # plt.ylim([-6,7])
    # plt.yticks([i for i in range(0, 1,5)])
    plt.ylabel("Population Growth")
    plt.xticks([i for i in range(2000, 2022,5)], rotation=45)
    plt.plot(Xs, ys[num], label = sea_countries[num])
    z = np.polyfit(Xs, ys[num], 1) # (X,Y,차원) 정의
    p = np.poly1d(z) # 1차원 다항식에 대한 연산을 캡슐화
#     pylab.plot(Xs,ys[num],'o')
    pylab.plot(Xs,p(Xs),"r--")
    plt.legend(title = "Legend Title")
    plt.show()
    print( "y=%.6fx+(%.6f)"%(z[0],z[1]))

    
# z = np.polyfit(Xs, ys[num], 1) # (X,Y,차원) 정의
# p = np.poly1d(z) # 1차원 다항식에 대한 연산을 캡슐화
    


# plt.show()
```
